````julia
#FUNGSI REGRESI LINEAR 
using Statistics
    function reglin(X,Y)
    xmean = mean(X);
    ymean = mean(Y);
    # Hitung nilai jumlah dari xy dan x^2
    sumxy = (X.-xmean)'*(Y.-ymean)
    sumx2 = (X.-xmean)'*(X.-xmean)
    # Hitung nilau koefisien garis regresi linear Y=Ax+B
    A = sumxy/sumx2;
    B = ymean .- A*xmean;
    return A,B
end;
````

````julia
#1a
x= [-6,-2,0,2,6]
y= [7,5,3,2,0]
A,B = reglin(x,y)

f(x) = A*x+B

using Plots
xk = -6:0.1:6
plot(xk,f.(xk),label = "Garis Regresi Linear")
scatter!(x,y,label = "Data a")
````

````julia
rmse(yk,yduga) = sqrt(mean((yk.-yduga).^2));
yduga = f.(x);
galat = rmse(y,yduga)
````

````julia
#FUNGSI REGRESI PANGKAT
function regpower(X,Y,m)
    sumxy = (X.^m)'*Y
    sumx2 = (X.^m)'*(X.^m)
    A = sumxy/sumx2
end;
````

````julia
#FUNGSI REGRESI POLINOMIAL 
function regpoly(X,Y,m)
    F = zeros(length(X),m+1)
    for k = 1: m+1
        F[:,k]=X.^(k-1);
    end
    A = F'*F;
    B = F'*Y;
    C = A\B;
    C = reverse(C,dims=1)
end;
````

#### 3a dan 3b

````julia
using Plots
using Statistics

# Data Pertama
xk_a = [-1, 0, 1, 2, 3]
yk_a = [6.62, 3.94, 2.17, 1.35, 0.89]

# Data Kedua
xk_b = [-3.0, -1.5, 0.0, 1.5, 3.0]
yk_b = [-0.14, -2.16, 0.83, 2.28, -0.51]

# Regresi Linear
function linear_regression(x, y)
    n = length(x)
    x̄ = mean(x)
    ȳ = mean(y)
    Σxy = sum(x .* y)
    Σx² = sum(x .^ 2)
    m = (n * Σxy - sum(x) * sum(y)) / (n * Σx² - sum(x) ^ 2)
    c = ȳ - m * x̄
    return m, c
end

# Nilai E2 (f)
function e2(y, ŷ)
    n = length(y)
    return sum((y .- ŷ) .^ 2) / n
end

# Plot hasil regresi
function plot_regression(x, y, m, c, label)
    scatter(x, y, label="Data")
    plot!(x, x -> m * x + c, label=label)
end

# Regresi untuk Data Pertama (a)
m_a, c_a = linear_regression(xk_a, yk_a)
e2_a = e2(yk_a, m_a * xk_a .+ c_a)

# Regresi untuk Data Kedua (b)
m_b, c_b = linear_regression(xk_b, yk_b)
e2_b = e2(yk_b, m_b * xk_b .+ c_b)

# Plot hasil regresi dalam satu grafik
plot_regression(xk_a, yk_a, m_a, c_a, "Regresi Data Pertama (a)")
plot_regression(xk_b, yk_b, m_b, c_b, "Regresi Data Kedua (b)")
xlabel!("xk")
ylabel!("yk")
title!("Hasil Regresi Linear untuk Kedua Set Data")
display(plot!())

println("Nilai E2 (f) untuk Data Pertama (a): ", e2_a)
println("Nilai E2 (f) untuk Data Kedua (b): ", e2_b)
````

````julia
using DataFrames
using CSV 
using LinearModels
using StatsPlots

# Membaca data dari file .txt
df = CSV.read("https://physics.info/curve-fitting/computer-temperatures.txt", DataFrame)

# Memisahkan kolom running time dan suhu komponen
running_time = df[:, :Running_time]
cpu_a = df[:, :CPU_A]
cpu_b = df[:, :CPU_B]
memory_controller_heatsink = df[:, :Memory_Controller_Heatsink]
drive_bay = df[:, :Drive_Bay]
hard_drive = df[:, :Hard_Drive]

# Membuat persamaan regresi linier untuk masing-masing komponen
cpu_a_model = lm(cpu_a ~ running_time)
cpu_b_model = lm(cpu_b ~ running_time)
memory_controller_heatsink_model = lm(memory_controller_heatsink ~ running_time)
drive_bay_model = lm(drive_bay ~ running_time)
hard_drive_model = lm(hard_drive ~ running_time)

# Menampilkan persamaan regresi dan grafik
println("Persamaan Regresi CPU A: CPU_A = $(coef(cpu_a_model)[1]) + $(coef(cpu_a_model)[2]) * Running_time")
println("Persamaan Regresi CPU B: CPU_B = $(coef(cpu_b_model)[1]) + $(coef(cpu_b_model)[2]) * Running_time")
println("Persamaan Regresi Memory Controller Heatsink: Memory_Controller_Heatsink = $(coef(memory_controller_heatsink_model)[1]) + $(coef(memory_controller_heatsink_model)[2]) * Running_time")
println("Persamaan Regresi Drive Bay: Drive_Bay = $(coef(drive_bay_model)[1]) + $(coef(drive_bay_model)[2]) * Running_time")
println("Persamaan Regresi Hard Drive: Hard_Drive = $(coef(hard_drive_model)[1]) + $(coef(hard_drive_model)[2]) * Running_time")

scatter(running_time, cpu_a, label="CPU A")
scatter!(running_time, cpu_b, label="CPU B")
scatter!(running_time, memory_controller_heatsink, label="Memory Controller Heatsink")
scatter!(running_time, drive_bay, label="Drive Bay")
scatter!(running_time, hard_drive, label="Hard Drive")
plot!(xlabel="Running time (menit)", ylabel="Suhu (°C)", legend=:top)
````
