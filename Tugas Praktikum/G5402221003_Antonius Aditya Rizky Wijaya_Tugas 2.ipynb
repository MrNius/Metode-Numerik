#### __Soal 1__
__Galat Mutlak dan Relatif.__ Hitunglah nilai error $E_{x}$ dan $R_{x}$ jika diketahui :

__(a)__ $x = 2.71828182$, $xˆ = 2.7182$

__(b)__ $y = 98350$, $xˆ = 98000$

__(c)__ $z = 0.000068$, $xˆ = 0.00006$

````julia
function ex_galat_mutlak(x, xˆ)
    return abs(x - xˆ)
end;

function rx_galat_relatif(x, xˆ)
    return abs(x - xˆ) / abs(x)
end;
````

````julia
# (a)
xa = 2.71828182
xˆa = 2.7182
Exa = ex_galat_mutlak(xa, xˆa)
Rxa = rx_galat_relatif(xa, xˆa)

println("(a)\nGalat Mutlak (Ex) = ", Exa, "\nGalat Relatif (Rx) = ", Rxa)
````

````julia
# (b)
xb = 98350
xˆb = 98000
Exb = ex_galat_mutlak(xb, xˆb)
Rxb = rx_galat_relatif(xb, xˆb)

println("(b)\nGalat Mutlak (Ex) = ", Exb, "\nGalat Relatif (Rx) = ", Rxb)
````

````julia
# (c)
xc = 0.000068
xˆc = 0.00006
Exc = ex_galat_mutlak(xc, xˆc)
Rxc = rx_galat_relatif(xc, xˆc)
println("(c)\nGalat Mutlak (Ex) = ", Exc, "\nGalat Relatif (Rx) = ", Rxc)
````

#### __Soal 2__
__Significant Digit.__ Jika $a = −0.00048914$, dan $b = 9125800%$ serta $x = 31.415$, dan $y = 0.027182$.

__(a)__ Hitung $a+b$ dan $ab$ dengan menggunakan 4 angka penting.

__(b)__ Hitung $x+y$ dan $xy$ dengan menggunakan 5 angka penting.

````julia
function pembulatan(x,n)
  m = log10(abs(x));
  if m > 0
    m = ceil(m);
    y = round(x .* 10 .^ (n-m)) ./ 10 .^ (n-m);
  else
    m = ceil(-m);
    y = round(x .* 10 .^ (m+n-1)) ./ 10 .^ (m+n-1);
  end
end;
````

````julia
a = -0.00048914;
b = 9125800;
x = 31.415;
y = 0.027182;
````

````julia
aplusb_4 = pembulatan(pembulatan(a,4)+pembulatan(b,4),4)
ab_4 = pembulatan(pembulatan(a,4)*pembulatan(b,4),4)
println("a + b = ", aplusb_4)
println("ab = ", ab_4)
````

````julia
xplusy_5 = pembulatan(pembulatan(x,5)+pembulatan(y,5),5)
xy_5 = pembulatan(pembulatan(x,5)*pembulatan(y,5),5)
println("x + y = ", xplusy_5)
println("xy = ", xy_5)
````

#### __Soal 3__
Diketahui fungsi $f$ dan $g$ sebagai berikut:
$$f(x) = \sqrt{x^2 + 1} - x$$
$$g(x) = \frac{1}{\sqrt{x^2 + 1} + x}$$

__(a)__ Buatlah fungsi $f$ dan $g$ tersebut dengan nama $y = f x(x)$ dan $y = gx(x)$

__(b)__ Buatlah fungsi $fn(x,n)$ dan $gn(x,n)$ yaitu memunculkan nilai fungsi $f$ dan $g$ dengan kepresisian terbatas (menggunakan $n$ angka desimal).

__(c)__ Bila fungsi $fx$ dan $gx$ diasumsikan memberikan nilai eksak dari fungsi $f$ dan $g$, buatlah fungsi yang mengeluarkan nilai galat relatif dari $fn$ dan $gn$, yaitu

$$ E_f(x,n)=\left| \dfrac{fn(x,n)-fx(x)}{fx(x)}\right| \ \ \ dan \ \ \ E_g(x,n)=\left| \dfrac{gn(x,n)-gx(x)}{gx(x)}\right| $$

__(d)__ Gambarkan fungsi $E_{f}$ dan $E_{g}$ pada selang $x : 0 ≤ x ≤ 1000$ dengan sub-interval sebesar $0.05$ menggunakan $n = 7$ angka penting.

__(e)__ Jelaskan hasil yang didapatkan dan kaitannya dengan _loss of significant digit._

````julia
# (a)
fx(x) = sqrt(x^2 + 1) - x;
gx(x) = 1 ./ (sqrt(x^2 + 1) + x);
````

````julia
# (b)
function pembulatan(x,n)
  m = log10(abs(x));
  if m > 0
    m = ceil(m);
    y = round(x .* 10 .^ (n-m)) ./ 10 .^ (n-m);
  else
    m = ceil(-m);
    y = round(x .* 10 .^ (m+n-1)) ./ 10 .^ (m+n-1);
  end
end;

fn(x,n) = pembulatan(sqrt(x^2 + 1),n) - pembulatan(x,n);
gn(x,n) = 1 ./ (pembulatan(sqrt(x^2 + 1),n) + pembulatan(x,n));
````

````julia
# (c)
Ef(x,n) = abs.((fn.(x,n)-fx.(x))./fx.(x));
Eg(x,n) = abs.((gn.(x,n)-gx.(x))./gx.(x));
````

````julia
#(d)

using Plots
function plot_Ef_Eg(n)
    selang = 0:0.05:1000
    nilai_Ef = [Ef(x, n) for x in selang]
    nilai_Eg = [Eg(x, n) for x in selang]
    plot(selang, nilai_Ef, label="Ef(x,$n)", xlabel="x", ylabel="Galat Relatif", legend=:topleft)
    plot!(selang, nilai_Eg, label="Eg(x,$n)")
end;
plot_Ef_Eg(7)
````

__# (e)__

Berdasarkan plot di atas, sepanjang selang $x : 0 ≤ x ≤ 1000$ fungsi $g$ memiliki galat relatif yang lebih kecil (malah terlihat sangat kecil) dibanding dengan fungsi $f$. Hal ini akibat dari terjadinya _lost of significant digit_ yang disebabkan oleh pengurangan dua angka yang saling berdekatan pada fungsi $f$, yaitu $\sqrt{x^2+1}$ dan $x$ untuk $x$ bernilai besar. Sehingga meskipun secara analitik fungsi $f$ dan fungsi $g$ menghasilkan nilai yang sama, tetapi dalam proses komputasi lebih baik menggunakan fungsi $g$ dibanding fungsi $f$.

#### __Soal 4__

Konversi bilangan biner berikut ke bilangan desimal.

__(a)__ $10101_{2}$

__(b)__ $111000_{2}$

__(c)__ $111111110_{2}$

__(d)__ $1000000111_{2}$

````julia
function convert_2_to_10(biner)
    desimal = 0
    for (i, digit) in enumerate(reverse(biner))
        desimal += digit * 2^(i-1)
    end
    return desimal
end;
````

````julia
# (a)
a_2 = [1, 0, 1, 0, 1]
a_10 = convert_2_to_10(a_2)
println("(a)\nBiner: 10101\nDesimal: ", a_10, "\n")

# (b)
b_2 = [1, 1, 1, 0, 0, 0]
b_10 = convert_2_to_10(b_2)
println("(b)\nBiner: 111000\nDesimal: ", b_10, "\n")

# (c)
c_2 = [1, 1, 1, 1, 1, 1, 1, 1, 0]
c_10 = convert_2_to_10(c_2)
println("(c)\nBiner: 111111110\nDesimal: ", c_10, "\n")

# (d)
d_2 = [1, 0, 0, 0, 0, 0, 0, 1, 1, 1]
d_10 = convert_2_to_10(d_2)
println("(d)\nBiner: 1000000111\nDesimal: ", d_10)
````

### __Soal 7__

Konversi bilangan pecahan desimal berikut ke biner 

__(a)__ $1/10$

__(b)__ $1/3$

__(c)__ $1/7$

````julia
function convert_10_to_2_koma(desimal, bit=16)
    biner_koma = []
    while desimal != 0.0 && length(biner_koma) < bit
        desimal *= 2
        digit = Int(floor(desimal))
        push!(biner_koma, digit)
        desimal -= digit
    end
    return biner_koma
end;
````

````julia
# (a)
biner_koma_a = convert_10_to_2_koma(1/10)
println("(a)\nBiner: 1/10\nDesimal: 0.", join(string(digit) for digit in biner_koma_a), "\n")

# (b)
biner_koma_b = convert_10_to_2_koma(1/3)
println("(b)\nBiner: 1/3\nDesimal: 0.", join(string(digit) for digit in biner_koma_b), "\n")

# (c)
biner_koma_c = convert_10_to_2_koma(1/7)
println("(c)\nBiner: 1/7\nDesimal: 0.", join(string(digit) for digit in biner_koma_c))
````

#### __Soal 8__

Hitung beberapa persamaan berikut, jika diketahui sebuah komputer hanya mampu menyimpan 4-bit mantissa.

__(a)__ $(\frac{1}{3}+\frac{1}{5}) + \frac{1}{6}$

__(b)__ $(\frac{1}{10}+\frac{1}{3}) + \frac{1}{5}$

__(c)__ $(\frac{3}{17}+\frac{1}{9}) + \frac{1}{7}$

__(d)__ $(\frac{7}{10}+\frac{1}{9}) + \frac{1}{7}$

````julia
function bulat_4bit(x)
    pembulatan = round(x; digits=4)
    return pembulatan
end;
````

````julia
a1 = bulat_4bit(1/3)
a2 = bulat_4bit(1/5)
a3 = bulat_4bit(1/6)
(a) = (a1 + a2) + a3

b1 = bulat_4bit(1/10)
b2 = bulat_4bit(1/3)
b3 = bulat_4bit(1/5)
(b) = (b1 + b2) + b3

c1 = bulat_4bit(3/17)
c2 = bulat_4bit(1/9)
c3 = bulat_4bit(1/7)
(c) = (c1 + c2) + c3

d1 = bulat_4bit(7/10)
d2 = bulat_4bit(1/9)
d3 = bulat_4bit(1/7)
(d) = (d1 + d2) + d3

println("(a) (1/3 + 1/5) + 1/6 = ", (a))
println("(b) (1/10 + 1/3) + 1/5 = ", (b))
println("(c) (3/17 + 1/9) + 1/7  = ", (c))
println("(d) (7/10 + 1/9) + 1/7 = ", (d))
````
