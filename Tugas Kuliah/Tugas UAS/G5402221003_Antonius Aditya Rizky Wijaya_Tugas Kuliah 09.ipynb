### 6.3
Use a library routine, or one of your own design, to find a minimum of each of the following functions on the interval $[0, 3]$. Draw a plot of each function to confirm that it is unimodal.

````julia
using Pkg
Pkg.add("Plots")
````

````julia
using Plots
function goldensection(f,a,b; tol = 1e-8)
    τ = (sqrt(5)-1)/2
    x1 = a+(1-τ)*(b-a);f1=f(x1)
    x2 = a+τ*(b-a);f2=f(x2)
    M = [x1 f1 x2 f2]
    iter = 1
    while b-a > tol && iter<1000
        if f1 > f2
            a=x1
            x1=x2;f1=f2
            x2=a+τ*(b-a);f2=f(x2)
        else
            b=x2
            x2=x1;f2=f1
            x1=a+(1-τ)*(b-a);f1=f(x1)
        end
        M = [M; x1 f1 x2 f2]
        iter += 1
    end
    return x1,M
end;
````

````julia
using Plots
function newton(f, df, ddf, x0; tol = 1e-8)
    x1 = x0 - df(x0)/ddf(x0)
    M = [x0 f(x0);x1 f(x1)]
    iter = 1
    while abs(x1-x0)>tol && iter<1000
        x0 = x1
        x1 = x0 - df(x0)/ddf(x0)
        M = [M; x1 f(x1)]
        iter += 1
    end
    return x1,M
end;
````

#### (a) $f(x) = x^4 − 14x^3 + 60x^2 − 70x$

##### Metode Golden Section

````julia
f(x) = x^4 - 14x^3 + 60x^2 - 70x;
min_x, M1 = goldensection(f,0,3)
println("Nilai minimum dari fungsi (a) : ", f(min_x))
println("Nilai x yang menghasilkan nilai minimum : ", min_x, "\n")

nilai_x = range(0, stop=3, length=100)
nilai_y = f.(nilai_x)
plot(nilai_x, nilai_y, label="x^4 - 14x^3 + 60x^2 - 70x", xlabel="x", ylabel="f(x)", title="Plot f(x) = x^4 - 14x^3 + 60x^2 - 70x")
scatter!([M1[:, 1]], [M1[:, 2]], label="Iterasi Golden Section", color=:red, markersize=2)
````

##### Metode Newton

````julia
f(x) = x^4 - 14x^3 + 60x^2 - 70x;
df(x) = 4x^3 - 42x^2 + 120x - 70;
ddf(x) = 12x^2 - 84x + 120;
min_x, M1 = newton(f, df, ddf, 1.0)
println("Nilai minimum dari fungsi (a) : ", f(min_x))
println("Nilai x yang menghasilkan nilai minimum : ", min_x, "\n")

nilai_x = range(0, stop=3, length=100)
nilai_y = f.(nilai_x)
plot(nilai_x, nilai_y, label="x^4 - 14x^3 + 60x^2 - 70x", xlabel="x", ylabel="f(x)", title="Plot f(x) = x^4 - 14x^3 + 60x^2 - 70x")
scatter!([M1[:, 1]], [M1[:, 2]], label="Iterasi Newton", color=:red, markersize=2)
````

#### (b) $f(x) = 0.5x^2 − sin(x)$

##### Metode Golden Section

````julia
f(x) = 0.5x^2 - sin(x);
min_x, M2 = goldensection(f,0,3)
println("Nilai minimum dari fungsi (b) : ", f(min_x))
println("Nilai x yang menghasilkan nilai minimum : ", min_x, "\n")

nilai_x = range(0, stop=3, length=100)
nilai_y = f.(nilai_x)
plot(nilai_x, nilai_y, label="0.5x^2 - sin(x)", xlabel="x", ylabel="f(x)", title="Plot f(x) = 0.5x^2 - sin(x)")
scatter!([M2[:, 1]], [M2[:, 2]], label="Iterasi Golden Section", color=:red, markersize=2)
````

##### Metode Newton

````julia
f(x) = 0.5x^2 - sin(x);
df(x) = x - cos(x);
ddf(x) = 1 + sin(x);
min_x, M2 = newton(f, df, ddf, 1.0)
println("Nilai minimum dari fungsi (b) : ", f(min_x))
println("Nilai x yang menghasilkan nilai minimum : ", min_x, "\n")

nilai_x = range(0, stop=3, length=100)
nilai_y = f.(nilai_x)
plot(nilai_x, nilai_y, label="0.5x^2 - sin(x)", xlabel="x", ylabel="f(x)", title="Plot f(x) = 0.5x^2 - sin(x)")
scatter!([M2[:, 1]], [M2[:, 2]], label="Iterasi Newton", color=:red, markersize=2)
````

#### (c) $f(x) = x^2 + 4 cos(x)$

##### Metode Golden Section

````julia
f(x) = x^2 + 4cos(x);
min_x, M3 = goldensection(f,0,3)
println("Nilai minimum dari fungsi (c) : ", f(min_x))
println("Nilai x yang menghasilkan nilai minimum : ", min_x, "\n")

nilai_x = range(0, stop=3, length=100)
nilai_y = f.(nilai_x)
plot(nilai_x, nilai_y, label="x^2 + 4cos(x)", xlabel="x", ylabel="f(x)", title="Plot f(x) = x^2 + 4cos(x)")
scatter!([M3[:, 1]], [M3[:, 2]], label="Iterasi Golden Section", color=:red, markersize=2)
````

##### Metode Newton

````julia
f(x) = x^2 + 4cos(x);
df(x) = 2x - 4sin(x);
ddf(x) = 2 - 4cos(x);
min_x, M3 = newton(f, df, ddf, 2.0)
println("Nilai minimum dari fungsi (c) : ", f(min_x))
println("Nilai x yang menghasilkan nilai minimum : ", min_x, "\n")

nilai_x = range(0, stop=3, length=100)
nilai_y = f.(nilai_x)
plot(nilai_x, nilai_y, label="x^2 + 4cos(x)", xlabel="x", ylabel="f(x)", title="Plot f(x) = x^2 + 4cos(x)")
scatter!([M3[:, 1]], [M3[:, 2]], label="Iterasi Newton", color=:red, markersize=2)
````
